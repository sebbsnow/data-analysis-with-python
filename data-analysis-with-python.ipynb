```python
# Compter le nombre de valeur pour chaque réponse de la colonne (Serie) "What is your gender?" 
# en affichant aussi le nombre de valeurs nulles
data["What is your gender?"].value_counts(dropna=False)
```

```python
# Importer la librairie "math"
import math

# Transfomer les valeurs catéorielles de la colonne "What is your gender?" en valeurs numériques
def gender_code(gender_string):
    if isinstance(gender_string, float) and math.isnan(gender_string):
        return gender_string
    return int(gender_string == "Female")
```

```python
# Appliquer la fonction "gender_code" (voir ci-dessus) sur la colonne (Serie) "What is your gender?"
data["gender"] = data["What is your gender?"].apply(gender_code)

# Afficher le nombre de valeur pour chaque réponse de la colonne "What is your gender?" 
# en affichant aussi le nombre de valeurs nulles ("dropna=False")
data["gender"].value_counts(dropna=False)
```

## UTILISATION DE LA MÉTHODE "APPLY" POUR NETTOYER "INCOME"

```python
# Compter le nombre de valeur pour chaque réponse de la colonne 
# "How much total combined money did all members of your HOUSEHOLD earn last year?" 
# en affichant aussi le nombre de valeurs nulles
data["How much total combined money did all members of your HOUSEHOLD earn last year?"].value_counts(dropna=False)
```

```python
# Importer la librairie numpy
import numpy as np

# Créer la fonction "clean_income"
def clean_income(value):
    # Vérifiez si la value est égale ou supérieure à 200 000 $
    if value == "$200,000 and up":
        # Retourner 200,000
        return 200000
    # Vérifiez si la value est égale à "Prefer not to answer"
    elif value == "Prefer not to answer":
        # Retourner NaN
        return np.nan
    # Vérifiez si la value est NaN
    elif isinstance(value, float) and math.isnan(value):
        # Retourner NaN
        return np.nan
    # Nettoyez la value en supprimant tout signe dollar ou virgule
    value = value.replace(",", "").replace("$", "")
    # Divisez la chaîne pour extraire les revenus
    income_high, income_low = value.split(" to ")
    # Retourner la moyenne
    return (int(income_high) + int(income_low)) / 2
```

```python
# Appliquer (apply) la fonction "clean_income" à la Serie data["income"]
data["income"] = data["How much total combined money did all members of your HOUSEHOLD earn last year?"].apply(clean_income)
# Afficher les (5) premières lignes
data["income"].head()
```
