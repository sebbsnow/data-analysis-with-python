## INDEXING DATAFRAME

### INDEXING PAR POSITION

```python
# Afficher les (5) premières lignes du dataset
# 0:5 => correspond aux lignes de 0 à 5 (exclue)
#  :  => correspond à toutes les colonnes
reviews.iloc[0:5,:]
```

```python
# Remplacer le DataFrame "reviews" avec toutes les lignes et à partir de la colonne 1 du DataFrame "reviews"
reviews = reviews.iloc[:,1:] 
```

### INDEXING PAR LABEL

```python
# Afficher les (5) premières lignes du dataset
reviews.loc[0:5,:]
```

```python
# Les index ne correspondent pas toujours aux positions.
# Exemple en créant un nouveau DataFrame dans lequel les positions et labels ne sont pas les mêmes
# Récupérer les lignes 10 à 20 de "reviews" et assignez le résultat à "some_reviews" 
some_reviews = reviews.iloc[10:20,] 

# Affiche les 5 premières lignes de "some_reviews"
some_reviews.head()
```

```python
# Tenter d'afficher les lignes 9 à 20
# Cela va retourner une erreur
some_reviews.loc[9:21,:]
```

## INDEXATION BOOLÉENNE DANS PANDAS 

```python
# Créer un DataFrame "score_filter" et y enregistrer les valeurs booléennes de "reviews["score"]" > 7
score_filter = reviews["score"] > 7 

# Afficher la Serie "score_filter"
score_filter
```

```python
# Créer un DataFrame "filtered_reviews" et y insérer la Serie "score_filter"
filtered_reviews = reviews[score_filter]

# Afficher les (5) premières lignes du DataFrame "filtered_reviews"
filtered_reviews.head()
```

```python
# Créer un DataFrame "xbox_one_filter" 
# et y enregistrer les valeurs booléennes de compâraison
# (reviews["score"]> 7) & (reviews["platform"] == "Xbox One")
xbox_one_filter = (reviews["score"]> 7) & (reviews["platform"] == "Xbox One")

# Créer un DataFrame "filter_reviews" et y insérer la Serie "xbox_one_filter"
filter_reviews = reviews[xbox_one_filter]

# Afficher les (5) premières lignes du DataFrame
filter_reviews.head()
```

```python
# Fonction spéciale de Jupyter pour afficher les plots sur Jupyter Notebook
%matplotlib inline

# Tracer la Serie reviews["score"] filtrée par reviews["platform"] == "Xbox One"
# Traçage de type "hist"
reviews[reviews["platform"] == "Xbox One"]["score"].plot(kind="hist")
```
